The purpose of this document is to show how to use the testing API to run tests. The following example shows a simple test.

```python
START_DATE = date(2015, 1, 1)
END_DATE = date(2022, 12, 31)

demo_tickers = ["EURUSD-Z"]

STARTING_EQUITY = 1_000_000
ACCOUNT_CURRENCY = "USD"
```

1. Initialise MT5

```python
init_mt5(login=None, password=None, server="AdmiralMarkets-Demo")
configLogger(logging.ERROR)
```

2. Set Strategy

```python
from strategies.previous_day_trend_strategy import PreviousDayTrendStrategy

demo_strategy = PreviousDayTrendStrategy()
```

3. Create Account

```python
from art_trader.abstract.testing import BacktestAccount

demo_account = BacktestAccount(initial_balance=STARTING_EQUITY, currency=ACCOUNT_CURRENCY)
```

3. Create Backtester

```python
from art_trader.mt5.testing import MT5Backtester

bt = MT5Backtester(strategy=demo_strategy, tickers=demo_tickers, start=START_DATE, end=END_DATE, account=demo_account)
```

4. Run test

```python
result = bt.run_all_single_thread()
```

5. Plot the results

```python
result.plot(y="balance", title=f"Testing from {START_DATE} to {END_DATE},\ntickers={demo_tickers}", use_index=True)
pass
```
